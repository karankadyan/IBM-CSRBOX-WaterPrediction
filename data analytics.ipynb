# Contigency Table

```python
from scipy.stats import chi2_contingency

# Discretize 'chromium' for creating a contingency table
df['chromium_category'] = pd.cut(df['chromium'], bins=3, labels=['low', 'medium', 'high'])

# Create a contingency table
contingency_table = pd.crosstab(df['chromium_category'], df['is_safe'])
contingency_table
```

# Performing Chi-Square test

```python
from scipy.stats import chi2_contingency

# Perform chi-square test
chi2, p, dof, expected = chi2_contingency(contingency_table)

chi2, p, dof, expected
```

# Performing t-test

```python
import pandas as pd
from scipy.stats import ttest_1samp, ttest_ind

numeric_columns = ['aluminium', 'arsenic', 'barium', 'cadmium', 'chloramine', 'chromium', 'copper', 'flouride', 'bacteria', 'viruses', 'lead', 'nitrates', 'nitrites', 'mercury', 'perchlorate', 'radium', 'selenium', 'silver', 'uranium']

# One-sample t-test
for col in numeric_columns:
    t_stat, p_value = ttest_1samp(df[col], popmean=0)  # Replace 0 with your hypothesized value
    print(f"{col}: t-statistic = {t_stat:.4f}, p-value = {p_value:.4f}")
```
